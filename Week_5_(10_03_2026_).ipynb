```python
#Python Program to Compute Population Statistics, Sample Statistics, and Z-Score
import numpy as np
import math

# 1. Population data
population = np.array([10, 20, 30, 40, 50, 60, 70])

N = len(population)
print("Population size (N):", N)

# Population mean
mu = np.mean(population)
print("Population mean (μ):", mu)

# Population variance
sigma2 = np.var(population, ddof=0)
print("Population variance (σ²):", sigma2)

# Population standard deviation
sigma = np.sqrt(sigma2)
print("Population standard deviation (σ):", sigma)

# 2. Sample data
sample = np.array([20, 30, 40, 50])

n = len(sample)
print("\nSample size (n):", n)

# Sample mean
x_bar = np.mean(sample)
print("Sample mean:", x_bar)

# Sample variance
s2 = np.var(sample, ddof=1)
print("Sample variance:", s2)

# Sample standard deviation
s = np.sqrt(s2)
print("Sample standard deviation:", s)

# 3. Z-score calculation for x = 50
x = 50
z = (x - mu) / sigma
print("\nZ-score for x=50:", z)
```

```python
# hypothesis code
import math

n = 100
x = 60

p_hat = x / n

SE = math.sqrt(p_hat * (1 - p_hat) / n)

z = 1.96

lower = p_hat - z * SE
upper = p_hat + z * SE

print("Sample proportion:", p_hat)
print("Confidence Interval:", (lower, upper))

p0 = 0.50

z_test = (p_hat - p0) / math.sqrt(p0 * (1 - p0) / n)

print("z value:", z_test)

if abs(z_test) > 1.96:
    print("Reject H0")
else:
    print("Do not Reject H0")
```

```python
#random variable model
import numpy as np

# Random variable data
X = np.array([10,20,30,40,50,60,70])

# Number of observations
n = len(X)
print("Total observations:", n)

# Probability of each value (discrete random variable)
p = 1/n

print("\nRandom Variable Values and Probabilities:")
for x in X:
    print("P(X =", x, ") =", p)

# Expected value E(X)
EX = np.sum(X * p)
print("\nExpected Value E(X):", EX)

# Variance Var(X)
variance = np.sum(((X - EX)**2) * p)
print("Variance Var(X):", variance)

# Standard Deviation
std = np.sqrt(variance)
print("Standard Deviation:", std)
```

```python
#central limit theorem
import numpy as np

# Population data
population = np.array([10,20,30,40,50,60,70])

# Population mean
mu = np.mean(population)
print("Population Mean:", mu)

# Population standard deviation
sigma = np.std(population)
print("Population Standard Deviation:", sigma)

# Number of samples
num_samples = 1000
sample_size = 3

sample_means = []

# Central Limit Theorem simulation
for i in range(num_samples):
    sample = np.random.choice(population, sample_size)
    sample_mean = np.mean(sample)
    sample_means.append(sample_mean)

# Mean of sample means
mean_sample_means = np.mean(sample_means)

print("Mean of Sample Means:", mean_sample_means)
```

```python
#standardizations
import numpy as np

# Data values
data = np.array([10,20,30,40,50,60,70])

# Mean of data
mean = np.mean(data)
print("Mean:", mean)

# Standard deviation
std = np.std(data)
print("Standard Deviation:", std)

# Standardization (Z-score)
z_scores = (data - mean) / std

print("\nStandardized Values (Z-scores):")
for i in range(len(data)):
    print("Value:", data[i], " Z-score:", z_scores[i])
```
