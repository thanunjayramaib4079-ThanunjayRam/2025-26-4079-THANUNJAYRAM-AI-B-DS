```python
# Creating tuples
t1 = (10, 20, 30, 40)
t2 = (50, 60)

# Accessing elements
print(t1[0])
print(t1[-1])

# Slicing
print(t1[1:3])
print(t1[::-1])

# Concatenation
t3 = t1 + t2
print(t3)

# Repetition
t4 = t1 * 2
print(t4)

# Membership
print(20 in t1)
print(100 not in t1)

# Length
print(len(t1))

# Iteration
for i in t1:
    print(i)

# Tuple methods
print(t1.count(20))
print(t1.index(30))
```

```python
# Creating sets
s1 = {10, 20, 30}
s2 = set([30, 40, 50])

# Adding elements
s1.add(40)

# Removing elements
s1.remove(20)    # error if not found
s1.discard(100)  # no error if not found

# Membership test
print(30 in s1)
print(100 not in s1)

# Set operations
print(s1.union(s2))          # union
print(s1.intersection(s2))   # intersection
print(s1.difference(s2))     # difference
print(s1.symmetric_difference(s2))  # elements in either but not both

# Length
print(len(s1))

# Iteration
for elem in s1:
    print(elem)
```

```python
# Creating dictionaries
d = {"name": "Alice", "age": 25, "city": "NY"}

# Access values
print(d["name"])
print(d.get("age"))

# Add / update
d["email"] = "alice@example.com"
d["age"] = 26

# Remove
del d["city"]
d.pop("email")

# Keys, values, items
print(d.keys())
print(d.values())
print(d.items())

# Membership (keys)
print("name" in d)
print("city" not in d)

# Iteration over keys and values
for key in d:
    print(key, d[key])
```
